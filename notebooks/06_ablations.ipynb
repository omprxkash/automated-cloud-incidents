# 06 — Ablation Studies
Ablate: graph k-value, patch size, and edge weight threshold τ.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ── Ablation 1: k-value vs accuracy (placeholder — fill with actual run results) ──
# These values should come from running train.py with different --graph-k values
# and reading the resulting metrics.txt files.

k_vals   = [1, 2, 3, 4, 6, 8]
top1_acc = [35.2, 39.8, 41.5, 43.1, 42.7, 42.6]  # placeholder

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(k_vals, top1_acc, 'o-', color='steelblue', linewidth=2, markersize=8)
ax.axvline(4, color='red', linestyle='--', alpha=0.7, label='k=4 (chosen)')
ax.set_xlabel('k (graph neighbours)')
ax.set_ylabel('Top-1 Accuracy (%)')
ax.set_title('Ablation: Graph k-Value vs Accuracy')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/ablation_k_value.png', dpi=120)
plt.show()
```

```python
# ── Ablation 2: patch_k (feature map patch size) ──
pk_vals  = [2, 4, 8]
top1_acc2 = [40.1, 43.1, 41.8]  # placeholder

fig, ax = plt.subplots(figsize=(7, 5))
ax.bar([str(p) for p in pk_vals], top1_acc2, color=['#aaa', 'steelblue', '#aaa'])
ax.set_xlabel('Patch Size k (feature-map pixels)')
ax.set_ylabel('Top-1 Accuracy (%)')
ax.set_title('Ablation: Feature-Map Patch Size')
ax.set_ylim(35, 48)
for i, v in enumerate(top1_acc2):
    ax.text(i, v + 0.2, f'{v:.1f}%', ha='center')
plt.tight_layout()
plt.savefig('../results/ablation_patch_size.png', dpi=120)
plt.show()
```

```python
# ── Ablation 3: threshold τ ──
tau_vals = [0.0, 0.05, 0.1, 0.2, 0.4]
top1_acc3 = [42.5, 43.0, 43.1, 42.3, 40.8]  # placeholder

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(tau_vals, top1_acc3, 's--', color='coral', linewidth=2, markersize=8)
ax.axvline(0.1, color='green', linestyle='--', alpha=0.7, label='τ=0.1 (chosen)')
ax.set_xlabel('Edge Weight Threshold τ')
ax.set_ylabel('Top-1 Accuracy (%)')
ax.set_title('Ablation: Sparse Graph Pruning Threshold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/ablation_threshold.png', dpi=120)
plt.show()
```
