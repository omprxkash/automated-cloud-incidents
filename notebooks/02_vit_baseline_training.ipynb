```python
import torch
from src.models import ViTSmall

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Device: {device}')

model = ViTSmall(img_size=128, patch_size=16, num_classes=100).to(device)
n = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'ViT-Small params: {n/1e6:.2f}M')
```

```python
# Launch training via CLI (recommended for Colab)
# !python ../src/train.py --model vit --epochs 100 --batch 128 --out ../results

# Or run a quick 2-epoch smoke test inline:
import subprocess
result = subprocess.run(
    ['python', '../src/train.py', '--model', 'vit', '--epochs', '2',
     '--batch', '64', '--out', '../results', '--data', '../data'],
    capture_output=True, text=True
)
print(result.stdout[-3000:])  # last 3k chars
print(result.stderr[-500:])
```

```python
# Plot training curves after full run
from src.visualize import plot_training_curves
from pathlib import Path

log_path = '../results/vit/train_log.csv'
if Path(log_path).exists():
    plot_training_curves(log_path, Path('../results/vit'))
    from IPython.display import Image
    display(Image('../results/vit/training_curves.png'))
else:
    print('Run training first.')
```

```python
# Evaluate checkpoint
import subprocess
ckpt = '../results/vit/best.pt'
if Path(ckpt).exists():
    r = subprocess.run(
        ['python', '../src/evaluate.py', '--model', 'vit',
         '--checkpoint', ckpt, '--data', '../data', '--out', '../results'],
        capture_output=True, text=True
    )
    print(r.stdout)
else:
    print('Checkpoint not found — run training first.')
```
